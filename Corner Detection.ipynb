```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### CORNER  is a point whose local neighbourhood stands in two dominant and different edge directions

## Most common Corner Detection Algo:

### 1. Harris Corner Detection
### 2. Shi-Tomasi Corner Detection

### Shi-Tomasi Algo is a small modification in Harris Corner Detection Algo

### Shi-Tomasi uses a different 'scoring function selection' criteria 

```python
img = cv2.imread('flat_chessboard.png')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img)
```

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

```python
plt.imshow(img_gray)
```

```python
plt.imshow(img_gray,cmap='gray')
```

```python
img_gray
```

```python
img_gray = np.float32(img_gray)
img_gray
```

```python
corner_1 = cv2.cornerHarris(src=img_gray, blockSize=2,ksize=3,k=0.04)
```

```python
corner_1 = cv2.dilate(corner_1,None)
```

```python
img[corner_1>0.01*corner_1.max()] = [255,0,0]
```

```python
plt.imshow(img)
```

## Shi-Tomasi Algo

```python
img_3 = cv2.imread('flat_chessboard.png')
img_3 = cv2.cvtColor(img_3, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img_3)
```

```python
plt.imshow(img_gray)
```

```python
corner_3 = cv2.goodFeaturesToTrack(image = img_gray,maxCorners = 100,qualityLevel=0.02,minDistance=10)
```

```python
corner_3
```

```python
corner_3 = np.int8(corner_3)
```

```python
corner_3
```

```python
for i in corner_3:
    x,y = i.ravel()
    cv2.circle(img_3, (x,y),3,(255,0,0),-1)
```

```python
plt.imshow(img_3)
```

```python
cv2.imshow('Image Corners', img_3)
cv2.waitKey()
cv2.destroyAllWindows()
```
