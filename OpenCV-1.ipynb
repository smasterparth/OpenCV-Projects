```python
new_img = cv2.imread('dog.jpg', cv2.IMREAD_GRAYSCALE)

while True:
    
    cv2.imshow('image', new_img)
    
    # If we waited for atleast 1ms & we have pressed ESC key, then BREAK.
    if cv2.waitKey(1) & 0xff == 27:
        break
        
cv2.destroyAllWindows()
```
