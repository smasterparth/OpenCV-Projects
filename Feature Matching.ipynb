```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### FEATURE MATCHING extracts defining key features from I/P image and then using a distance calculation, finds all the matches in a secondary image.

### We no longer need an exact copy of target image. 

## Following methods to be used for FEATURE MATCHING

#### 1. Brute-Force Matching with ORB Decriptors
#### 2.Brute-Force Matching with SIFT Descriptors & Ratio Test
#### 3. FLANN based Matcher

```python
def disp(img):
    fig = plt.figure(figsize=(8,8))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
```

```python
recess = cv2.imread('reeses_puffs.png',0)
```

```python
disp(recess)
```

```python
cereals = cv2.imread('many_cereals.jpg',0)
```

```python
disp(cereals)
```

```python
orb = cv2.ORB_create()
```

```python
kp1,des1 = orb.detectAndCompute(recess,None)
```

```python
kp2,des2 = orb.detectAndCompute(cereals,None)
```

```python
bf = cv2.BFMatcher(cv2.NORM_HAMMING,crossCheck=True)
```

```python
matches = bf.match(des1,des2)
```

```python
matches
```

```python
match_1 = matches[0]
match_1
```

```python
match_1.distance
```

```python
matches = sorted(matches,key = lambda x:x.distance)
```

```python
matches
```

```python
recess_matches = cv2.drawMatches(recess,kp1,cereals,kp2,matches[:30],None,flags=2)
```

```python
disp(recess_matches)
```
