```python
import cv2
import numpy as np
```

```python
from pyzbar.pyzbar import decode
```

```python
cap = cv2.VideoCapture(0)
cap.set(3,640)
cap.set(4,480)

while True:

    ret,image = cap.read()

    for i in decode(image):
        myData = i.data.decode('utf-8')
        print(myData)
        pts = np.array([i.polygon], np.int32)
        pts = pts.reshape((-1,1,2))
        cv2.polylines(image,[pts],True,(255,0,255),4)
        pts2 = i.rect
        cv2.putText(image,myData,(pts2[0],pts2[1]), cv2.FONT_HERSHEY_COMPLEX,0.5,(255,0,255),1)

    cv2.imshow('Result',image)
    
    key = cv2.waitKey(1)
    
    if key == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
